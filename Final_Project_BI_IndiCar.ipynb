#Setting Up Data for Usage

```python
#drop column yang tidak diperlukan
columns_to_drop = ['Id', 'inv_car_price', 'inv_car_dist', 'inv_car_age', 'inv_brand',
                   'std_invprice', 'std_invdistance_travelled', 'std_invrank']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
# tambah variable 'distance >30km'
data['distance >30k km'] = data['distance_travelled(kms)'] > 30000

# tambah variable 'old and more used'
data['old and more used'] = (data['year'] <= 2016) & (data['distance_travelled(kms)'] > 30000)

# rename column distance below 30k km
data.rename(columns={'distance below 30k km': 'distance <30k km'}, inplace=True)
```

```python
new_column_order = [
    'year', 'brand', 'full_model_name', 'model_name', 'price', 'distance_travelled(kms)', 'fuel_type',
    'city', 'brand_rank', 'car_age', 'distance <30k km', 'new and less used',
    'distance >30k km', 'old and more used', 'best_buy1', 'best_buy2'
    # Include all other columns you want to keep in the desired order
]

# Reorder the DataFrame columns
data = data[new_column_order]
```

```python
data['distance >30k km'] = data['distance >30k km'].astype(int)
data['old and more used'] = data['old and more used'].astype(int)
```

## Regression Analysis

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = data[['car_age', 'distance_travelled(kms)']]
y = data['price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Predict using the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("R-squared:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```

```python
X_age = data[['car_age']]  # Predictor: Age of Car
y = data['price']  # Response Variable: Price

# Separate regression for 'Distance Travelled'
X_distance = data[['distance_travelled(kms)']]  # Predictor: Distance Travelled

# Splitting data for 'Age of Car'
X_train_age, X_test_age, y_train, y_test = train_test_split(X_age, y, test_size=0.3, random_state=42)

# Splitting data for 'Distance Travelled'
X_train_distance, X_test_distance = train_test_split(X_distance, test_size=0.3, random_state=42)

# Initialize the model
model = LinearRegression()

# Train and evaluate the model for 'Age of Car'
model.fit(X_train_age, y_train)
y_pred_age = model.predict(X_test_age)
print("Age of Car -> R-squared:", r2_score(y_test, y_pred_age))
print("Age of Car -> MSE:", mean_squared_error(y_test, y_pred_age))

# Train and evaluate the model for 'Distance Travelled'
model.fit(X_train_distance, y_train)
y_pred_distance = model.predict(X_test_distance)
print("Distance Travelled -> R-squared:", r2_score(y_test, y_pred_distance))
print("Distance Travelled -> MSE:", mean_squared_error(y_test, y_pred_distance))

# Plotting results for 'Age of Car'
plt.figure(figsize=(10, 5))
plt.scatter(X_test_age, y_test, color='blue', label='Actual')
plt.scatter(X_test_age, y_pred_age, color='red', label='Predicted')
plt.title('Actual vs Predicted Prices: Age of Car')
plt.xlabel('Age of Car')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plotting results for 'Distance Travelled'
plt.figure(figsize=(10, 5))
plt.scatter(X_test_distance, y_test, color='blue', label='Actual')
plt.scatter(X_test_distance, y_pred_distance, color='red', label='Predicted')
plt.title('Actual vs Predicted Prices: Distance Travelled')
plt.xlabel('Distance Travelled')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## Cluster Analysis

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
features = data[['brand', 'price', 'city']]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['price']),
        ('cat', OneHotEncoder(), ['brand', 'city'])
    ])
```

```python
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('cluster', KMeans(n_clusters=5, random_state=42))])  # Adjust n_clusters based on your analysis
```

```python
pipeline.fit(features)
```

```python
labels = pipeline.named_steps['cluster'].labels_

# Attach cluster labels to your original DataFrame
data['Cluster'] = labels
```

```python
numeric_data = data.select_dtypes(include=[np.number])  # Select only numeric columns

# Then, include your 'Cluster' column for grouping
numeric_data['Cluster'] = data['Cluster']

# Now perform the groupby operation
cluster_means = numeric_data.groupby('Cluster').mean()
```
